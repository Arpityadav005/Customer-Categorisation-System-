```python
# Column	                       Meaning
# ID	                    Unique identifier for each customer.
# Year_Birth	            Year of birth of the customer.
# Education	                Education level of the customer (e.g., Basic, High School, Graduate, Master, PhD).
# Marital_Status	        Marital status of the customer (e.g., Single, Married, Divorced, Widowed).
# Income	                Annual income of the customer (likely in local currency).
# Kidhome	                Number of children living at home.
# Teenhome	                Number of teenagers living at home.
# Dt_Customer	          Date when the customer joined/registered (Customer registration date).
# Recency	               Number of days since the customer last made a purchase.
# MntWines	                 Amount spent on wine products by the customer.
# MntFruits	Amount           spent on fruit products by the customer.
# MntMeatProducts	Amount spent on meat products.
# MntFishProducts	Amount spent on fish products.
# MntSweetProducts	Amount spent on sweet products (chocolates, candies).
# MntGoldProds	Amount spent on gold or premium products.
# NumDealsPurchases	Number of purchases made using a deal or promotion.
# NumWebPurchases	Number of purchases made through the company’s website.
# NumCatalogPurchases	Number of purchases made using a catalog.
# NumStorePurchases	Number of purchases made in physical stores.
# NumWebVisitsMonth	Number of visits to the company website in the last month.
# AcceptedCmp3	Whether the customer accepted campaign 3 (1 = yes, 0 = no).
# AcceptedCmp4	Whether the customer accepted campaign 4.
# AcceptedCmp5	Whether the customer accepted campaign 5.
# AcceptedCmp1	Whether the customer accepted campaign 1.
# AcceptedCmp2	Whether the customer accepted campaign 2.
# Complain	Whether the customer registered a complaint (1 = yes, 0 = no).
# Z_CostContact	Cost incurred to contact the customer (internal marketing metric).
# Z_Revenue	Revenue generated from the customer (internal metric).
# Response	Whether the customer responded to the last marketing campaign (1 = yes, 0 = no).
```
